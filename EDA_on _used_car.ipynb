# Correlation Heatmap

Visualize relationships between numerical features.

```python
plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

# Histogram

Understand data distribution.

```python
plt.figure(figsize=(8,5))

plt.hist(df["Price"],bins=30)

plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

# Seaborn Histogram

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Price"],kde=True)

plt.show()
```

# Detect Outliers

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Price"])

plt.show()
```

# Count Plot

Count Categories

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df,x="Fuel_Type")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df,x="Transmission")

plt.show()
```

# Bar Plot

Average Price by Brand

```python
plt.figure(figsize=(12,5))

sns.barplot(data=df,
            x="Brand",
            y="Price")

plt.xticks(rotation=90)

plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(8,6))

sns.scatterplot(data=df,
                x="Horsepower",
                y="Price")

plt.show()
```

# Line Plot
Average Price by Year

```python
year_price=df.groupby("Year")["Price"].mean().reset_index()

plt.figure(figsize=(10,5))

sns.lineplot(data=year_price,
             x="Year",
             y="Price")

plt.show()
```

# Pair Plot

Relationship among numerical variables.

```python
sns.pairplot(df[[
    "Price",
    "Horsepower",
    "Engine_CC",
    "Mileage_kmpl",
    "Kms_Driven"
]])
```

# Voilin Plot

```python
plt.figure(figsize=(8,5))

sns.violinplot(data=df,
               x="Fuel_Type",
               y="Price")

plt.show()
```

# KDE Plot

```python
plt.figure(figsize=(8,5))

sns.kdeplot(df["Price"],fill=True)

plt.show()
```

# Distribution Plot

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Mileage_kmpl"],kde=True)

plt.show()
```

# Pie Chart
Fuel Type Percentage

```python
fuel=df["Fuel_Type"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(fuel,
        labels=fuel.index,
        autopct="%1.1f%%")

plt.title("Fuel Type Distribution")

plt.show()
```
